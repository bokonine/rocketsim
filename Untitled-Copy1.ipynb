The circumference and area of a arbitrary polar function $f(\theta)$ are:
$$circumference = \int_{0}^{2 \pi} \sqrt{f(\theta)^2 + (\frac{d}{d\theta}f(\theta))^2} d\theta$$
$$area = \int_{0}^{2 \pi} \frac{1}{2}f(\theta)^2$$

```python
import numpy as np
from scipy import integrate
from scipy import misc
import matplotlib.pyplot as plt
import math
plt.rcParams['figure.figsize'] = [10, 10]

Do = 15
do = 3

r = lambda θ,dmm=0 : do/2+dmm+((1.5+1.5*np.cos(8*θ))*np.power(.95,dmm))
r_area = lambda θ,mm=0: .5*(r(θ,mm)**2)
hole_area = lambda b_mm: integrate.quad(r_area, 0, 2*math.pi, args=(b_mm))

drdθ = lambda θ: misc.derivative(r, θ)
r_circum = lambda θ,b_mm: np.sqrt(r(θ)**2 + drdθ(θ)**2)
hole_circumference = lambda b_mm: integrate.quad(r_circum, 0, 2*math.pi, args=(b_mm))
```

```python
# show burn progression
%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 15]

Δburn = np.arange(0,Do/2-do/2,1)
Δtheta = np.linspace(0,2*math.pi,1000)

ax = plt.subplot(111, projection='polar',label='1')
for burnInc in Δburn: ax.plot(Δtheta, r(Δtheta, burnInc))
ax.set_rmax(Do/2)
ax.set_rticks(np.arange(0,Do/2,1))  # less radial ticks
ax.set_rlabel_position(-22.5)  # get radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')

plt.show()

```
