The circumference and area of a arbitrary polar function $f(\theta)$ are:
$$circumference = \int_{0}^{2 \pi} \sqrt{f(\theta)^2 + (\frac{d}{d\theta}f(\theta))^2} d\theta$$
$$area = \int_{0}^{2 \pi} \frac{1}{2}f(\theta)^2$$

```python
import numpy as np
from scipy import integrate
from scipy import misc
import matplotlib.pyplot as plt
import math
import ipynb.fs.full.utils as utils

plt.rcParams['figure.figsize'] = [10, 10]

Do = 15
do = 2

r = lambda theta,dmm=0 : do/2+dmm+(np.abs(0.5*np.cos(3*theta)))
```

```python
#show burn progression
plt.rcParams['figure.figsize'] = [15, 15]

deltaburn = np.arange(0,Do/2-do/2,.5)
deltatheta = np.linspace(0,2*math.pi,1024)

ax = plt.subplot(111, projection='polar',label='1')
for burnInc in deltaburn: ax.plot(deltatheta, r(deltatheta, burnInc))
ax.set_rmax(Do/2)
ax.set_rticks(np.arange(0,Do/2,1))  # less radial ticks
ax.set_rlabel_position(-22.5)  # get radial labels away from plotted line
ax.grid(True)

plt.show()

```
